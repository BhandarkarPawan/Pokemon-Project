```python
from pyhive import hive
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
cols = ['abilities', 'against_bug', 'against_dark', 'against_dragon',
        'against_electric', 'against_fairy', 'against_fight', 'against_fire',
        'against_flying', 'against_ghost', 'against_grass', 'against_ground',
        'against_ice', 'against_normal', 'against_poison', 'against_psychic',
        'against_rock', 'against_steel', 'against_water', 'attack',
        'base_egg_steps', 'base_happiness', 'base_total', 'capture_rate',
        'classfication', 'defense', 'experience_growth', 'height_m', 'hp',
        'japanese_name', 'name', 'percentage_male', 'pokedex_number',
        'sp_attack', 'sp_defense', 'speed', 'type1', 'type2', 'weight_kg',
        'generation', 'is_legendary']
```

```python
def get_data(cursor):
    result = []
    while(True):
        try:
            result.append((cursor.next()))
        except StopIteration:
            print("Loading done")
            break
    return pd.DataFrame(result, columns=cols)
```

```python
cursor = hive.connect('localhost').cursor()

cursor.execute("DROP table Pokemon")

cursor.execute("create table pokemon (\
Abilities string, against_bug FLOAT, against_dark FLOAT, against_dragon FLOAT,\
against_electric FLOAT, against_fairy FLOAT, against_fight FLOAT, against_fire FLOAT, \
against_flying FLOAT, against_ghost FLOAT, against_grass FLOAT, against_ground FLOAT,\
against_ice FLOAT, against_normal FLOAT, against_poison FLOAT,  against_psychic FLOAT,\
against_rock FLOAT, against_steel FLOAT, against_water FLOAT, attack int,\
base_egg_steps int, base_happiness int, base_total int, capture_rate string,\
classfication string, defense int, experience_growth int, height_m FLOAT, hp int,\
japanese_name string, name string , percentage_male FLOAT, pokedex_number int,\
sp_attack int, sp_defense int, speed int, type1 string, type2 string, weight_kg FLOAT,\
generation int , is_legendary int) row format delimited fields terminated by ','")

cursor.execute("load data local inpath 'pokemon.txt' into table Pokemon")
```

```python
# View the first 5 rows of the DataSet
cursor.execute("Select * from pokemon LIMIT 5")
get_data(cursor)
```

```python
# View the last 5 rows of the DataSet
cursor.execute("Select * FROM pokemon ORDER BY Name DESC LIMIT 5")
get_data(cursor)
```

```python
# Set properties for visualizations 
plt.style.use('seaborn-pastel')
%matplotlib inline
plt.rcParams['figure.dpi'] = 100
```

```python
# Get all data from the database for cleaning purposes:
cursor.execute("SELECT * FROM pokemon")
pokemon = get_data(cursor)
```

```python
# Standard data cleaning procedures
pokemon.columns = pokemon.columns.str.capitalize()
pokemon.drop_duplicates(keep=False,inplace=True)
pokemon['Type2'].fillna(value='None', inplace=True)
pokemon['Weight_kg'].fillna(value='None', inplace=True)
pokemon['Percentage_male'].fillna(value='None', inplace=True)
pokemon['Height_m'].fillna(value='None', inplace=True)
pokemon['Type1'] = pokemon['Type1'].str.capitalize()
pokemon['Type2'] = pokemon['Type2'].str.capitalize()
```

```python
sns.catplot(
    x='Generation', 
    data=pokemon,
    height=3,
    aspect=1.2,
    kind='count'
).set_axis_labels('Generation', 'Number of Pokemons')

plt.show()
```

```python
# Using plots to visualize the distribution of Legendary and Non Legendary Pokemon 


cursor.execute("SELECT * FROM pokemon WHERE is_legendary == 1")
pokeLeg = get_data(cursor)
cursor.execute("SELECT * FROM pokemon WHERE is_legendary == 0")
pokeNon = get_data(cursor)

fig = plt.figure(figsize=(5,5))
colours = ["aqua", "gray"]
legDist = [pokeLeg['name'].count(),pokeNon['name'].count()]
legPie = plt.pie(legDist,
                 labels= ['Legendary', 'Non Legendary'], 
                 autopct ='%1.1f%%', 
                 shadow = False,
                 colors=colours,
                 startangle = 45,
                 explode=(0, 0.1))
```

```python
#Let us view how the legendary pokemon are distributed per generation 

plt.style.use('classic')
%matplotlib inline
plt.rcParams['figure.dpi'] = 100

colours = ["aqua", "gray"]
g = sns.catplot(
    x='Generation', 
    data=pokemon,
    kind='count', 
    hue='Is_legendary',
    palette=colours, 
    height=3.5, 
    aspect=1.5,
    legend=False,
    ).set_axis_labels('Generation', 'Number of Pokemons')

g.ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.1),  shadow=True, ncol=2, labels=['NON LEGENDARY','LEGENDARY'])

plt.show()
```

```python
# Distribution of pokemon types helps us identify the most common and rare types 

plt.style.use('fivethirtyeight')
%matplotlib inline
plt.rcParams['figure.dpi'] = 100
fig = plt.figure(figsize=(20,20))

fig.add_subplot(211)
pokemon['Type1'].value_counts().plot(kind='pie', 
                                       autopct='%1.1f%%',
                                       pctdistance=0.9)

# Consider only the pokemon that HAVE a Type2
fig.add_subplot(212)
pokemon[pokemon.Type2 != '']['Type2'].value_counts().plot(kind='pie', 
                                       autopct='%1.1f%%',
                                       pctdistance=0.9)

plt.show()
```

```python
plt.style.use('seaborn-pastel')
%matplotlib inline
plt.rcParams['figure.dpi'] = 100
fig = plt.figure(figsize=(20,20))

sns.catplot(
    y='Type1',
    data=pokemon,
    kind='count',
    order=pokemon['Type1'].value_counts().index,
    height=4,
    aspect=1.5,
    color='teal'
).set_axis_labels('Number of Pokemons', 'Type 1')

sns.catplot(
    y='Type2',
    data=pokemon,
    kind='count',
    order=pokemon['Type2'].value_counts().index,
    height=4,
    aspect=1.5,
    color='turquoise'
).set_axis_labels('Number of Pokemons', 'Type 2');
```

```python
# This plot will help us find which combination of Types is most common
plt.subplots(figsize=(10, 10))

sns.heatmap(
    pokemon[pokemon['Type2']!='None'].groupby(['Type1', 'Type2']).size().unstack(),
    linewidths=1,
    annot=True,
    cmap="Blues"
)

plt.xticks(rotation=35)
plt.show()
```
