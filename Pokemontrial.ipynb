```python
import pandas as pd   
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(plt.style.available)
```

```python
plt.style.use('seaborn-pastel')
%matplotlib inline
plt.rcParams['figure.dpi'] = 100
```

```python
pokemon = pd.read_csv("pokemon.csv")
```

```python
pokemon
```

```python
pokemon.head()  
```

```python
pokemon.tail()
```

```python
pokemon.columns = pokemon.columns.str.capitalize()
```

```python
pokemon
```

```python
pokemon.drop_duplicates(keep=False,inplace=True)
```

```python
pokemon
```

```python
pokemon['Type2'].fillna(value='None', inplace=True)
pokemon['Weight_kg'].fillna(value='None', inplace=True)
pokemon['Percentage_male'].fillna(value='None', inplace=True)
pokemon['Height_m'].fillna(value='None', inplace=True)
```

```python
pokemon
```

```python
pokemon['Type1'] = pokemon['Type1'].str.capitalize()
pokemon['Type2'] = pokemon['Type2'].str.capitalize()
pokemon 
```

```python
pokemon.count()
```

```python
sns.catplot(
    x='Generation', 
    data=pokemon,
    height=3,
    aspect=1.2,
    kind='count'
).set_axis_labels('Generation', 'Number of Pokemons')

plt.show()
```

```python
fig = plt.figure(figsize=(5,5))

colours = ["gray", "aqua"]
pokeLeg = pokemon[pokemon['Is_legendary']==True]
pokeNon = pokemon[pokemon['Is_legendary']==False]

legDist = [pokeLeg['Name'].count(),pokeNon['Name'].count()]
legPie = plt.pie(legDist,
                 labels= ['Legendary', 'Non Legendary'], 
                 autopct ='%1.1f%%', 
                 shadow = False,
                 colors=colours,
                 startangle = 45,
                 explode=(0, 0.1))
```

```python

plt.style.use('classic')
%matplotlib inline
plt.rcParams['figure.dpi'] = 100

colours = ["aqua", "gray"]
g = sns.catplot(
    x='Generation', 
    data=pokemon,
    kind='count', 
    hue='Is_legendary',
    palette=colours, 
    height=3.5, 
    aspect=1.5,
    legend=False,
    ).set_axis_labels('Generation', 'Number of Pokemons')

g.ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.1),  shadow=True, ncol=2, labels=['NON LEGENDARY','LEGENDARY'])

plt.show()
```

```python
plt.style.use('fivethirtyeight')
%matplotlib inline
plt.rcParams['figure.dpi'] = 100
fig = plt.figure(figsize=(20,20))

fig.add_subplot(211)
pokemon['Type1'].value_counts().plot(kind='pie', 
                                       autopct='%1.1f%%',
                                       pctdistance=0.9)

fig.add_subplot(212)
pokemon['Type2'].value_counts().plot(kind='pie', 
                                       autopct='%1.1f%%',
                                       pctdistance=0.9)

plt.show()
```

```python
plt.style.use('seaborn-pastel')
%matplotlib inline
plt.rcParams['figure.dpi'] = 100
fig = plt.figure(figsize=(20,20))

sns.catplot(
    y='Type1',
    data=pokemon,
    kind='count',
    order=pokemon['Type1'].value_counts().index,
    height=4,
    aspect=1.5,
    color='teal'
).set_axis_labels('Number of Pokemons', 'Type 1')

sns.catplot(
    y='Type2',
    data=pokemon,
    kind='count',
    order=pokemon['Type2'].value_counts().index,
    height=4,
    aspect=1.5,
    color='turquoise'
).set_axis_labels('Number of Pokemons', 'Type 2');
```

```python
plt.subplots(figsize=(10, 10))

sns.heatmap(
    pokemon[pokemon['Type2']!='None'].groupby(['Type1', 'Type2']).size().unstack(),
    linewidths=1,
    annot=True,
    cmap="Blues"
)

plt.xticks(rotation=35)
plt.show()
```
